```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sigma = 1
mu = 3

sample_100 = np.random.normal(mu, sigma, 100)
sample_10000 = np.random.normal(mu, sigma, 10000)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(sample_100, bins=20, density=True, color='b')
plt.title('Histogram dla 100')

plt.subplot(1, 2, 2)
plt.hist(sample_10000, bins=20, density=True, color='r')
plt.title('Histogram dla 10 000')


plt.tight_layout()
plt.show()
```

```python
#średnia
mean_sample_100 = np.round(np.mean(sample_100))
mean_sample_10000 = np.round(np.mean(sample_10000))
print(f"Średnia dla 100: {mean_sample_100}")
print(f"Średnia dla 10000: {mean_sample_10000}")

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(sample_100, bins=20)
plt.axvline(x = mean_sample_100, color='red', label = 'Średnia')
plt.legend(loc='upper right')
plt.title("Zmienna 100")

plt.subplot(1, 2, 2)
plt.hist(sample_10000, bins=20)
plt.axvline(x = mean_sample_10000, color='red', label = 'Średnia')
plt.legend(loc='upper right')
plt.title("Zmienna 10000")

plt.tight_layout()
plt.show()
```

```python
#mediana
median_sample_100 = np.round(np.median(sample_100))
median_sample_10000 = np.round(np.median(sample_10000))
print(f"Mediana dla 100: {median_sample_100}")
print(f"Mediana dla 10000: {median_sample_10000}")

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(sample_100, bins=20)
plt.axvline(x = mean_sample_100, color='red', label = 'Średnia')
plt.axvline(x = median_sample_100, color='green', linestyle='dotted', label = 'Mediana')
plt.legend(loc='upper right')
plt.title("Zmienna 100")

plt.subplot(1, 2, 2)
plt.hist(sample_10000, bins=20)
plt.axvline(x = mean_sample_10000, color='red', label = 'Średnia')
plt.axvline(x = median_sample_10000, color='green', linestyle='dotted', label = 'Mediana')
plt.legend(loc='upper right')
plt.title("Zmienna 10000")

plt.tight_layout()
plt.show()
```

```python
#kwartyle
q0_sample_100 = np.quantile(sample_100, 0.0)
q1_sample_100 = np.round(np.quantile(sample_100, 0.25))
q3_sample_100 = np.round(np.quantile(sample_100, 0.75))
q4_sample_100 = np.round(np.quantile(sample_100, 1.0))
print(f'Kwartyle dla 100:')
print(f'Q0: {q0_sample_100}')
print(f'Q1: {q1_sample_100}')
print(f'Q2: {median_sample_100}')
print(f'Q3: {q3_sample_100}')
print(f'Q4: {q4_sample_100}')

q0_sample_10000 = np.quantile(sample_10000, 0.0)
q1_sample_10000 = np.round(np.quantile(sample_10000, 0.25))
q3_sample_10000 = np.round(np.quantile(sample_10000, 0.75))
q4_sample_10000 = np.round(np.quantile(sample_10000, 1.0))
print(f'\nKwartyle dla 10000:')
print(f'Q0: {q0_sample_10000}')
print(f'Q1: {q1_sample_10000}')
print(f'Q2: {median_sample_10000}')
print(f'Q3: {q3_sample_10000}')
print(f'Q4: {q4_sample_10000}')


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(sample_100, bins=20)
plt.axvline(x = mean_sample_100, color='red', label = 'Średnia')
plt.axvline(x = median_sample_100, color='green', linestyle='dotted', label = 'Mediana')
plt.axvline(x = q0_sample_100, color='black', label = 'Min')
plt.axvline(x = q1_sample_100, color='black', label = 'Q1')
plt.axvline(x = q3_sample_100, color='black', label = 'Q3', linewidth = 0.5)
plt.axvline(x = q4_sample_100, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.title("Zmienna 100")

plt.subplot(1, 2, 2)
plt.hist(sample_10000, bins=20)
plt.axvline(x = mean_sample_10000, color='red', label = 'Średnia')
plt.axvline(x = median_sample_10000, color='green', linestyle='dotted', label = 'Mediana')
plt.axvline(x = q0_sample_10000, color='black', label = 'Min')
plt.axvline(x = q1_sample_10000, color='black', label = 'Q1')
plt.axvline(x = q3_sample_10000, color='black', label = 'Q3')
plt.axvline(x = q4_sample_10000, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.title("Zmienna 10000")

plt.tight_layout()
plt.show()
```

```python
#zakres
range_sample_100 =np.round(max(sample_100) - min(sample_100))
range_sample_10000 =np.round(max(sample_10000) - min(sample_10000))
print(f'Zakres dla 100: {range_sample_100}')
print(f'Zakres dla 10000: {range_sample_10000}')
```

```python
#rozstęp międzykwartylowy - IQR
iqr_sample_100 = q3_sample_100 - q1_sample_100
iqr_sample_10000 = q3_sample_10000 - q1_sample_10000
print(f'IQR dla 100: {iqr_sample_100}')
print(f'IQR dla 10000: {iqr_sample_10000}')

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(sample_100)
plt.title('Zmienna 100')

plt.subplot(1, 2, 2)
plt.boxplot(sample_10000)
plt.title('Zmienna 10000')

plt.tight_layout()
plt.show()
```

```python
#wariancja
variance_sample_100 = np.var(sample_100, ddof=1)
variance_sample_10000 = np.var(sample_10000, ddof=1)
print(f'Wariancja dla 100: {variance_sample_100}')
print(f'Wariancja dla 10000: {variance_sample_10000}')
```
