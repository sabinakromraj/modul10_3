```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sigma = 1
mu = 3

sample_100 = np.random.normal(mu, sigma, 100)
sample_10000 = np.random.normal(mu, sigma, 10000)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(sample_100, bins=20, density=True, color='b')
plt.title('Histogram dla 100')

plt.subplot(1, 2, 2)
plt.hist(sample_10000, bins=20, density=True, color='r')
plt.title('Histogram dla 10 000')

plt.tight_layout()
plt.show()
```

```python
print(f"Różnica pomiędzy statystykami dla sample_100 i sample_10000 wynika z rozmiaru próbek i wpływa na precyzję oraz wiarygodność statystyk. Próbka sample_10000 jest większa niż sample_100, co oznacza, że ma większą precyzję. Statystyki obliczone na podstawie większej próbki zwykle są dokładniejsze i bardziej zbliżone do wartości teoretycznych.")
```

```python
#średnia
mean_sample_100 = np.mean(sample_100)
mean_sample_10000 = np.mean(sample_10000)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(sample_100, bins=20)
plt.axvline(x = mean_sample_100, color='red', label = 'Średnia')
plt.legend(loc='upper right')
plt.title("Zmienna 100")

plt.subplot(1, 2, 2)
plt.hist(sample_10000, bins=20)
plt.axvline(x = mean_sample_10000, color='red', label = 'Średnia')
plt.legend(loc='upper right')
plt.title("Zmienna 10000")

plt.tight_layout()
plt.show()

print(f"Średnia (mean) dla sample_100 i sample_10000 wynosi odpowiednio: {mean_sample_100} i {mean_sample_10000}. Różnica między tymi dwiema średnimi wynika z rozmiaru próbek. W przypadku sample_10000, większa liczba próbek ma wpływ na bardziej precyzyjne oszacowanie średniej, co prowadzi do wyniku bardziej zbliżonego do teoretycznej wartości średniej rozkładu normalnego, która wynosi 3.")
```

```python
#mediana
median_sample_100 = np.median(sample_100)
median_sample_10000 = np.median(sample_10000)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(sample_100, bins=20)
plt.axvline(x = mean_sample_100, color='red', label = 'Średnia')
plt.axvline(x = median_sample_100, color='green', linestyle='dotted', label = 'Mediana')
plt.legend(loc='upper right')
plt.title("Zmienna 100")

plt.subplot(1, 2, 2)
plt.hist(sample_10000, bins=20)
plt.axvline(x = mean_sample_10000, color='red', label = 'Średnia')
plt.axvline(x = median_sample_10000, color='green', linestyle='dotted', label = 'Mediana')
plt.legend(loc='upper right')
plt.title("Zmienna 10000")

plt.tight_layout()
plt.show()

print(f"Mediana dla 100: {median_sample_100}\nMediana dla 10000: {median_sample_10000}\nRóżnica między tymi dwiema medianami wynika również z różnicy w rozmiarze próbek. Mediana sample_100 jest nieco wyższa niż mediana sample_10000. Jest to spowodowane przypadkowymi fluktuacjami w mniejszej próbce, które wpłynęły na wartość mediany.")
```

```python
#kwartyle
q0_sample_100 = np.quantile(sample_100, 0.0)
q1_sample_100 = np.quantile(sample_100, 0.25)
q3_sample_100 = np.quantile(sample_100, 0.75)
q4_sample_100 = np.quantile(sample_100, 1.0)

q0_sample_10000 = np.quantile(sample_10000, 0.0)
q1_sample_10000 = np.quantile(sample_10000, 0.25)
q3_sample_10000 = np.quantile(sample_10000, 0.75)
q4_sample_10000 = np.quantile(sample_10000, 1.0)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(sample_100, bins=20)
plt.axvline(x = mean_sample_100, color='red', label = 'Średnia')
plt.axvline(x = median_sample_100, color='green', linestyle='dotted', label = 'Mediana')
plt.axvline(x = q0_sample_100, color='black', label = 'Min')
plt.axvline(x = q1_sample_100, color='black', label = 'Q1')
plt.axvline(x = q3_sample_100, color='black', label = 'Q3')
plt.axvline(x = q4_sample_100, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.title("Zmienna 100")

plt.subplot(1, 2, 2)
plt.hist(sample_10000, bins=20)
plt.axvline(x = mean_sample_10000, color='red', label = 'Średnia')
plt.axvline(x = median_sample_10000, color='green', linestyle='dotted', label = 'Mediana')
plt.axvline(x = q0_sample_10000, color='black', label = 'Min')
plt.axvline(x = q1_sample_10000, color='black', label = 'Q1')
plt.axvline(x = q3_sample_10000, color='black', label = 'Q3')
plt.axvline(x = q4_sample_10000, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.title("Zmienna 10000")

plt.tight_layout()
plt.show()

print(f"Kwartyle dla 100:\nMin: {q0_sample_100}\nQ1: {q1_sample_100}\nQ2: {median_sample_100}\nQ3: {q3_sample_100}\nMax: {q4_sample_100}\nKwartyle dla 10000:\nMin: {q0_sample_10000}\nQ1: {q1_sample_10000}\nQ2: {median_sample_10000}\nQ3: {q3_sample_10000}\nMax: {q4_sample_10000}\nRóżnice między kwartylami dla sample_10000 są nieco niższe niż dla sample_100. Może to sugerować, że sample_10000 ma bardziej skoncentrowany rozkład danych wokół mediany, podczas gdy sample_100 jest bardziej zróżnicowane.")
```

```python
#zakres
range_sample_100 = max(sample_100) - min(sample_100)
range_sample_10000 = max(sample_10000) - min(sample_10000)
print(f'Zakres dla 100: {range_sample_100}\nZakres dla 10000: {range_sample_10000}\nZakres dla 10000 jest szerszy niż dla 100, co sugeruje większą zmienność w danych. Próbka o większym rozmiarze jest bardziej reprezentatywna dla rozkładu normalnego i ma tendencję do bardziej stabilnych statystyk, takich jak zakres.')
```

```python
#rozstęp międzykwartylowy - IQR
iqr_sample_100 = q3_sample_100 - q1_sample_100
iqr_sample_10000 = q3_sample_10000 - q1_sample_10000

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(sample_100)
plt.title('Zmienna 100')

plt.subplot(1, 2, 2)
plt.boxplot(sample_10000)
plt.title('Zmienna 10000')

plt.tight_layout()
plt.show()

print(f"IQR dla 100: {iqr_sample_100}\nIQR dla 10000: {iqr_sample_10000}\nRóżnica między zakresem międzykwartylowym dla 10000 jest nieco szersza niż dla 100, co sugeruje większą zmienność w danych w większej próbce.")
```

```python
#wariancja
variance_sample_100 = np.var(sample_100, ddof=1)
variance_sample_10000 = np.var(sample_10000, ddof=1)
print(f'Wariancja dla 100: {variance_sample_100}\nWariancja dla 10000: {variance_sample_10000}\nPróbka "100" zawierająca 100 wartości ma wariancję wynoszącą około 0.941, co oznacza, że dane w niej są mniej zmienne i skupione bliżej średniej. Natomiast próbka "10000" z 10 000 wartościami ma wyższą wariancję, wynoszącą około 1.012, co sugeruje większą zmienność danych i ich większe rozproszenie wokół średniej. Większa próbka generuje większą wariancję.')
```
